```
import pyspark
from pyspark import SparkConf, SparkContext
conf = SparkConf().setAppName("GroupReduce")
sc = SparkContext.getOrCreate(conf=conf)
```

```
rdd =sc.textFile("/FileStore/tables/groupreduce.txt")
```

```
rdd.collect()
```

```
rdd1 = rdd.flatMap(lambda x:x.split(" "))
```

```
rdd1.collect()
```

```
rdd2 = rdd1.map(lambda x:(x,1))
```

```
rdd2.collect()
```

```
rdd3 = rdd2.groupByKey().mapValues(len)
```

```
rdd3.collect()
```

```
rdd4 = rdd2.reduceByKey(lambda x,y : x+y)
```

```
rdd4.collect()
```
